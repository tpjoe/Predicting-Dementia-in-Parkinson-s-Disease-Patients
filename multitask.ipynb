# Machine Learning

## Conventional machine learning

```python
Xs = dict()
ys = dict()

for i in range(6):
    Xs[i] = pd.read_csv('data/X_yearsince1stvisit_stdScaled.{}.csv'.format(i)).iloc[:, 1:]
    ys[i] = pd.read_csv('data/y_yearsince1stvisit_stdScaled.{}.csv'.format(i), header=None).iloc[:, 1:]
```

```python
from sklearn.model_selection import StratifiedKFold, RepeatedKFold, GridSearchCV
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import roc_auc_score
from sklearn.linear_model import LogisticRegression

n_repeats = 25
n_splits = 4
rkf = RepeatedKFold(n_splits=n_splits, n_repeats=n_repeats)
# parameters = {'kernel':('linear', 'rbf'), 'C':[1, 10]}
# parameters = {'n_estimators':[50, 100, 500]}
parameters = {'n_neighbors':[1, 5]}
# parameters = {'penalty':['l1']}


auc_dict = {i:np.full([n_repeats*n_splits, 3], np.nan) for i in range(len(ys))}

for task in range(len(ys)):
    print('At task {}'.format(task))
    for i, (train_index, test_index) in enumerate(rkf.split(Xs[task], ys[task])):
        
        X_train, X_test = Xs[task].iloc[train_index, :], Xs[task].iloc[test_index, :]
        y_train, y_test = np.array(ys[task].iloc[train_index].transpose().values[0]), np.array(ys[task].iloc[test_index].transpose().values[0])
        
        if len(np.unique(y_test)) < 3:
            continue
        
#         clfer = SVC(probability=True)
#         clfer = RandomForestClassifier()
        clfer = KNeighborsClassifier()
#         clfer = LogisticRegression(solver='liblinear', max_iter=1000)
        clf = GridSearchCV(clfer, parameters, cv=3)
        clf.fit(X_train, y_train)
        y_pred = clf.predict_proba(X_test)
        
        for ii in range(3):
            auc_dict[task][i, ii] = roc_auc_score(pd.get_dummies(y_test).iloc[:, ii], y_pred[:, ii])
```

```python
# make a plot dataframe
task = [[i] * (n_repeats*n_splits*3) for i in range(len(ys))]
task = [j for i in task for j in i]
classes = [['PD-NCI', 'PD-MCI', 'PDD']*(n_repeats*n_splits*len(ys))]
classes = [j for i in classes for j in i]

plot_values = pd.DataFrame([np.vstack(list(auc_dict.values())).flatten(), task, classes]).transpose()
plot_values.columns = ['AUC', 'task', 'classes']
```

```python
# Plotting out
sns.reset_orig()
sns.set(font_scale=1.4)
sns.set_style("ticks", {"axes.facecolor": "1", 'axes.grid': True, 'axes.edgecolor': '.4'})
fig, ax = plt.subplots(1, 1, figsize=(9, 6))
boxPlot = sns.boxplot(x="task", y="AUC",
            hue="classes", palette=['#087e8b', '#f9dc5c', '#e84855'],
            data=plot_values, linewidth=0.85, fliersize=4, width=0.65)
ax.set_xlabel('Years since 1$^{st}$ visit')
ax.set_ylim([0, 1])
# replace labels
ax.legend_.remove()

# legends
NC = mlines.Line2D([], [], color='#087e8b', marker='s', linestyle='None', markersize=10, label='PD-NCI')
MCI = mlines.Line2D([], [], color='#f9dc5c', marker='s', linestyle='None', markersize=10, label='PD-MCI')
dementia = mlines.Line2D([], [], color='#e84855', marker='s', linestyle='None', markersize=10, label='PDD')
plt.legend(handles=[NC, MCI, dementia])
plt.title('K-nearest Neighbors')
# fig.savefig("figures/AUC_LASSO.svg", bbox_inches = "tight")
```

```python
y_pred = result['y_pred']
y_test = result['y_test']

values_dict = dict()
plot_values = pd.DataFrame()

for task in ['task1', 'task2', 'task3', 'task4', 'task5', 'task6']:
    prauc = np.zeros((3, 100))
    auc = np.zeros((3, 100))
    for iteration in range(0, 100):
        y_pred_0 = np.array(y_pred['no CI'][task])[:, iteration]
        y_pred_1 = np.array(y_pred['CI'][task])[:, iteration]
        y_pred_2 = np.array(y_pred['dementia'][task])[:, iteration]

        y = [y_pred_0, y_pred_1, y_pred_2]

        ground_truth = np.array(y_test[task])[:, iteration]

        for n in range(0, 3):
            gt = np.where(ground_truth == n, ground_truth, -1)
            prauc[n, iteration] = average_precision_score(gt, y[n], pos_label=n)
#             auc[n, iteration] = roc_auc_score(gt, y[n])

    AUCs_ = pd.DataFrame(prauc).apply(pd.Series).T.melt().rename(columns={'variable':'classes', 'value':'AUC'})
    AUCs_.classes = AUCs_.classes.apply(lambda x: ' No CI' if x==0 else 'CI' if x==1 else 'Dementia')
    values_dict[task] = AUCs_
    values_dict[task] = pd.concat([values_dict[task], pd.DataFrame([(task)]*values_dict[(task)].shape[0])], axis=1)

for task in values_dict:
    plot_values = pd.concat([plot_values, values_dict[task]], axis=0)
plot_values = plot_values.set_index('classes')
plot_values = plot_values.rename(columns={0:'task'})
# boxwhisker = hv.BoxWhisker(plot_values, ['task', 'classes'], 'AUC', label='AUC of each class by ' + result['clfs'] + ' model')

# boxwhisker = boxwhisker.opts(width=600, cmap=['#087e8b', '#f9dc5c', '#e84855'], box_fill_color=dim('classes').str(), xrotation = 30, ylabel="PRAUC").opts(
#     opts.BoxWhisker(show_legend=False, ylim=(0, 1), box_width=0.5, box_alpha=1, show_grid=True, box_line_width=0.25, 
#                     whisker_line_width=0.6, whisker_color='black', xlabel='Predicing visit (task no.)'))

# hv.output(boxwhisker, backend='bokeh', fig='html')
```

```python
# Compute ROC curve and ROC area for each class
from sklearn.metrics import roc_curve, auc
from sklearn.metrics import roc_auc_score

fpr = dict()
tpr = dict()
roc_auc = dict()
n_classes = 3
task = 'task4'
iteration = 95
for i, status in enumerate(['no CI', 'CI', 'dementia']):
    gr_truth = np.array(y_test[task])[:, iteration]
    gr_truth = [1 if j==i else 0 for j in gr_truth]
    fpr[i], tpr[i], _ = roc_curve(gr_truth, np.array(y_pred[status][task])[:, iteration])
    roc_auc[i] = auc(fpr[i], tpr[i])

plt.figure()
lw = 2
plt.plot(fpr[0], tpr[0], color='#087e8b',
         lw=lw, label='ROC curve (area = %0.2f)' % roc_auc[0])
plt.plot(fpr[1], tpr[1], color='#f9dc5c',
         lw=lw, label='ROC curve (area = %0.2f)' % roc_auc[1])
plt.plot(fpr[2], tpr[2], color='#e84855',
         lw=lw, label='ROC curve (area = %0.2f)' % roc_auc[2])
plt.plot([0, 1], [0, 1], color='grey', lw=lw, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC, Task 1, Iter 0')
plt.legend(loc="lower right")
plt.show()
```
